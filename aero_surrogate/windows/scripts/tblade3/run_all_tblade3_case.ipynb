```python
import subprocess
from pathlib import Path
import pandas as pd
```

```python
# ---------- CONFIG ----------

PROJECT_ROOT = Path(r"C:\Users\dipes\OneDrive\Documents\Machine Learning\Python\Structural Optimization\aero_surrogate")

DESIGNS_PATH = PROJECT_ROOT / "windows" / "datasets" / "designs_6dv.csv"
```

```python
# WSL paths (these are Linux-style paths passed to `wsl`)

WSL_SCRIPT = (
    "/mnt/c/Users/dipes/OneDrive/Documents/Machine Learning/"
    "Python/Structural Optimization/aero_surrogate/wsl/scripts/run_all_tblade3.sh"
)

WSL_FEACASES_ROOT = (
    "/mnt/c/Users/dipes/OneDrive/Documents/Machine Learning/"
    "Python/Structural Optimization/aero_surrogate/wsl/fea_cases"
)
```

```python
# ---------- LOAD DESIGN CASES ----------

df = pd.read_csv(DESIGNS_PATH)

if "case" not in df.columns:
    raise ValueError("designs_6dv.csv must have a 'case' column.")

case_ids = df["case"].astype(int).tolist()

print(f"Found {len(case_ids)} cases.")
print("First few cases:", case_ids[:10])
```

```python
# ---------- FUNCTION TO RUN ONE CASE ----------

def run_case(case_id: int):
    # Build WSL case directory path
    case_dir = f"{WSL_FEACASES_ROOT}/case_{case_id:04d}"

    cmd = ["wsl", WSL_SCRIPT, case_dir]

    print(f"[INFO] Running case {case_id:04d} ...")
    # If something goes wrong, check=True will raise an exception
    result = subprocess.run(cmd, capture_output=True, text=True)

    if result.returncode != 0:
        print(f"[ERROR] Case {case_id:04d} failed.")
        print("STDOUT:", result.stdout)
        print("STDERR:", result.stderr)
    else:
        print(f"[OK] Case {case_id:04d} completed.")

# ---------- LOOP OVER ALL CASES ----------

for cid in case_ids:
    run_case(cid)

print("All cases processed (as far as the script could run).")
```
